# Sveto naučno trojstvo

## Nikola Tesla

**Nikola Tesla**, čuveni srpski pronalazač i inženjer elektrotehnike i mašinstva,rođen je u smiljanu 10.jula 1856,a umro je u Njujoku,na *Božić 7.januara 1943*.

Pronalasci Nikole Tesle:
- polifazni sistem,
- obrtno magnetno polje,
- asinhroni motor,
- sinhroni motor,
- Teslin transformator.
